```python
import pandas as pd
```

```python
df = pd.read_csv('medical_examination.csv')
```

```python
df.head()
```

```python
df.loc[(df['weight']/(df['height']/100)**2) <= 25, 'overweight'] = 0
df.loc[(df['weight']/(df['height']/100)**2) > 25, 'overweight'] = 1
```

```python
df.head()
```

```python
df = df.astype({'overweight':int})
df
```

```python
df.loc[df['cholesterol'] == 1, 'cholesterol'] = 0
df.loc[df['cholesterol'] > 1, 'cholesterol'] = 1
df.loc[df['gluc'] == 1, 'gluc'] = 0
df.loc[df['gluc'] > 1, 'gluc'] = 1
df
```

```python
df_cat = pd.melt(frame=df, id_vars='cardio', value_vars=['active','alco','cholesterol','gluc','overweight','smoke'])
df_cat
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = sns.catplot(x='variable', data=df_cat, hue='value', col='cardio', kind='count')
fig.set(ylabel='total')
fig = fig.fig
```

```python
 df_heat = df[(df['ap_lo'] <= df['ap_hi']) &
    (df['height'] >= df['height'].quantile(0.025)) &
    (df['height'] <= df['height'].quantile(0.975)) &
    (df['weight'] >= df['weight'].quantile(0.025)) &
    (df['weight'] <= df['weight'].quantile(0.975))]
df_heat
```

```python
corr = df_heat.corr()
```

```python
mask = np.triu(corr)
mask
```

```python
fig, ax = plt.subplots(figsize=(11,9))
fig = sns.heatmap(data=corr, mask=mask, annot=True,vmax=0.3, center=0, square=True, fmt=".1f")
fig = fig.get_figure()
```
